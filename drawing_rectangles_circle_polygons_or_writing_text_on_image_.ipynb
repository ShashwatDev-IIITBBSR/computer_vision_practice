```python
#blankimage
blank_img=np.zeros(shape=(512,512,3),dtype=np.int16)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,(100,100),(300,300),thickness=10,color=(0,0,255))
```

```python
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,center=(400,400),thickness=5,radius=120,color=(0,255,255))
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(300,100),pt2=(400,400),thickness=5,color=(255,0,255))
plt.imshow(blank_img)
```

```python
#another blank image
blank_img2=np.zeros(shape=(1000,1000,3),dtype=np.int32)
```

```python
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(blank_img2,text='SHASHWAT DEV',org=(10,500),fontFace=font,fontScale=4,color=(0,255,255),thickness=5,lineType=cv2.LINE_AA)
plt.imshow(blank_img2)
```

```python
#POLYGONS
vertices=np.array([[100,300],[500,400],[500,300],[800,900]],dtype=np.int32)
pts=vertices.reshape((-1,1,2))
pts.shape
```

```python
cv2.polylines(blank_img2,[pts],isClosed=True,color=(255,255,0),thickness=6)
plt.imshow(blank_img2)
```
